```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
data = pd.read_csv("alt_fuel_stations (Jan 9 2024).csv")
data.head()
```

```python
# Creating a new data frame that only uses the data needed for the project 
df = data[['Fuel Type Code', 'City', 'State', 'Longitude', 'Latitude', 'Street Address', 'EV Network']]
print(df.sample(10))
```

### Creating a column for regions

```python
#creating array to store region data
west = ['WA', 'OR', 'ID', 'MT', 'WY', 'CA', 'NV', 'UT', 'CO', 'AK', 'HI']
southwest = ['AZ', 'NM', 'OK', 'TX']
southeast = ['AR', 'LA', 'MS', 'AL', 'GA', 'FL', 'SC', 'NC', 'TN', 'KY', 'VA', 'WV', 'MD', 'DE', 'PR' ]
northeast = ['PA', 'NJ', 'CT', 'RI', 'NY', 'VT', 'ME', 'NH', 'MA', 'DC']
midwest = ['ND', 'SD', 'NE', 'KS', 'MO', 'IA', 'MN', 'WI', 'MI','IL', 'IN', 'OH']

print(len(northeast) + len(southeast) + len(west) + len(southwest) + len(midwest))

#function that accepts state as input and maps according to array
def map_to_region(state):
    if state in west:
        return 'West'
    elif state in southwest:
        return 'Southwest'
    elif state in southeast:
        return 'Southeast'
    elif state in northeast:
        return 'Northeast'
    elif state in midwest:
        return 'Midwest'
    else:
        return 'Invalid'
    
#using .apply() to iterate through 
df['Region'] = df['State'].apply(map_to_region)
df.sample(10)
```

```python
df['Region'].value_counts()
```

```python
#Seeing where the invalid results are coming from
df.loc[df['Region'] == 'Invalid']
```

```python
#Removing these invalid results since it is unclear
df = df[df['Region'] != 'Invalid']
df['Region'].value_counts()
```

```python
#Checking for any missing values in columns

#Loops through each column of df and checks for missing values,
# prints a statement if there are

for column_name in df.columns:
    missing_count = df[column_name].isna().sum()
    if (missing_count > 0):
        print(f"In '{column_name}' there are {missing_count} missing values")
```

```python
#Filling the city with zero
df['City'].fillna('Unknown', inplace=True)
print(df['City'].isna().sum())

#Filling in the unknown street addresses
df['Street Address'].fillna('Unknown', inplace=True)
print(df['City'].isna().sum())
```

### Exporting to CSV

```python
df.to_csv('alt_fuel_stations_prepped.csv', index=False)
```
